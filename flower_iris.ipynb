```
import numpy as np 
```

```
import pandas as pd
```

```
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
new_names = ['sepal_length','sepal_width','petal_length','petal_width','iris_class']
data=pd.read_csv(url,names=new_names,skiprows=0)
data    #import data from url and change names of columns 
```

```
data.shape #to see the shape of data
```

```
data.isnull()
```

```
data.isnull().sum().any() # to see if any data is null or not to avoid error in our prediction
```

```
data.isnull().sum() #column wise check null values
```

```
import matplotlib.pyplot as plt
```

```
cols=['sepal_length','sepal_width','petal_length','petal_width'] # x- axis features
left=['iris_class']
```

```
x=data[cols]

y=data[left]
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) # split training dataset from test dataset
```

```
from sklearn.linear_model  import LogisticRegression
```

```
from sklearn import metrics 
```

```
from sklearn.metrics import accuracy_score
```

```
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```
print(accuracy_score(y_test,log_reg.predict(x_test))) # prediction always change as it depend upon the training dataset.
```

```
print(y_test,log_reg.predict(x_test))
```

```
from sklearn.svm import SVC 
svc=SVC()
svc.fit(x_train,y_train)  #using svm
```

```
print(accuracy_score(y_test,svc.predict(x_test)))
```
